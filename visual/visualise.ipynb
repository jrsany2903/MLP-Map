```python
import pyvista as pv

# Set interactive notebook backend
pv.set_jupyter_backend("trame")

# Create a sphere (Earth placeholder)
sphere = pv.Sphere(radius=1.0, theta_resolution=360, phi_resolution=180)

# Create the plotter
plotter = pv.Plotter()
plotter.add_mesh(sphere, color="lightblue")  # placeholder color

# Show interactively
plotter.show()
```
